```python
import glob
import urllib2
import matplotlib.pyplot as plt

from StringIO import StringIO
from PIL import Image

%matplotlib inline
plt.rcParams['figure.figsize'] = 24,12

img_locations = []
img_locations.append("http://www.spc.noaa.gov/climo/reports/130531_rpts.gif")
img_locations.append("http://www.spc.noaa.gov/climo/reports/110427_rpts.gif")
img_locations.append("http://www.spc.noaa.gov/climo/reports/120629_rpts.gif")
img_locations.append("http://www.spc.noaa.gov/climo/reports/100605_rpts.gif")

#loop through all file locations
for i in range(len(img_locations)):
    ##read in image
    img_data = urllib2.urlopen(img_locations[i]).read()
    img = Image.open(StringIO(img_data))
    ##set current subplot to i
    ##subplot shape will be in this case
    ##(2,2,i) where i is the current subplot
    plt.subplot(len(img_locations)/2,len(img_locations)/2,i)
    
    ##show image
    plt.imshow(img,vmin=0,vmax=16)
```

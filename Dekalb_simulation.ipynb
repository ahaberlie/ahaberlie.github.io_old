```python
import numpy as np
import matplotlib.pyplot as plt
import math
import time

from IPython.display import display, clear_output

%matplotlib inline

plt.rcParams['figure.figsize'] = 12,6

##500 by 500 km grid!
wx_grid = np.zeros((500,500),dtype=np.float)
wx_grid1 = np.zeros((500,500),dtype=np.float)

##set temps to random values in reasonable kelvin range
tmp_range = np.linspace(263,283,500)
for i in range(wx_grid.shape[0]):
    for j in range(wx_grid.shape[1]):
        wx_grid[j,i] = (tmp_range[j]+tmp_range[i])/ 2.

wx_grid = np.flipud(wx_grid)
wx_grid = np.fliplr(wx_grid)

##calculate temperature gradient
x_grad = (wx_grid[-1,0] - wx_grid[0,0])/ (float(wx_grid.shape[0])*1000)
y_grad = (wx_grid[0,-1] - wx_grid[0,0])/ (float(wx_grid.shape[1])*1000)

##set u and v winds 
u = 10.
v = -20.

##set up the grid for contouring
x = np.arange(0,wx_grid.shape[0], 1)
y = np.arange(0,wx_grid.shape[1], 1)
X, Y = np.meshgrid(x, y)
Z = wx_grid

f, ax = plt.subplots()
##DT/Dt = 0 obviously
for iteration in range(5000):
    time.sleep(1)
    
    ##compare this to nested iteration. OMG
    wx_grid += -u*x_grad - v*y_grad
    
    Z = wx_grid
    ax.contour(X,Y,Z,25,colors = ('k',),vmin=245,vmax=290)
    ax.contourf(X,Y,Z,25,vmin=245,vmax=290)
    ax.plot(300,300,'ko',markersize=15)
    plt.title("Temp at DeKalb: "+str((Z[300,300]-273.15)*1.8)+"F")
    ax.barbs(300,300, u, v,length=10)
    #f.colorbar(ax)
    display(f)
    clear_output(wait=True)
    ax.cla()
```

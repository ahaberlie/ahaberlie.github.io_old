```python
import glob
import matplotlib.pyplot as plt

#I use this for reading images.. prefer it over PIL but your mileage may vary
from scipy.ndimage import imread

%matplotlib inline
plt.rcParams['figure.figsize'] = 12,6

#the *.gif at the end gets all the gifs from the directory and ignores everything else
fileDir = "your/directory/here/*.gif"

#here is the file location list
d = glob.glob(fileDir)

#loop through all file locations
for i in range(len(d)):
    ##read in image
    img = imread(d[i])
    
    ##set current subplot to i
    ##subplot shape will be in this case
    ##(2,2,i) where i is the current subplot
    plt.subplot(len(d)/2,len(d)/2,i)
    
    ##show image
    plt.imshow(img,vmin=0,vmax=16)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

from mpl_toolkits.basemap import Basemap
from scipy.spatial import cKDTree as tree
from scipy.ndimage import gaussian_filter as gf

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 10)

##my attempt at creating grids based on x and y resolution
##still a work in progress...
def get_grid(x_resolution,y_resolution,lon0,lat0,lon1,lat1):
    x_bins = abs(lon1-lon0) * ((111.2*math.sin(lon1) - 111.2*math.sin(lon0)) / 2) / y_resolution
    y_bins = (lat1-lat0) * 111.2 / y_resolution
    xi = np.linspace(-130,-60,y_bins)
    yi = np.linspace(20,50,x_bins)
    return xi,yi
```

```python
##example data.. can be any lat,lon data
torData = pd.read_csv('https://raw.githubusercontent.com/ahaberlie/hovmoller/master/torn_reports_minimal.csv',delimiter=',',header=0)

x = torData['LON']
x = x[x<-60]
y = torData['LAT']
y = y[y>15]

llcrnrlon = -120
llcrnrlat = 20
urcrnrlon = -60
urcrnrlat = 50
m = Basemap(projection='lcc', lat_0 = 39, lon_0 = -96, lat_1 = 40,
              llcrnrlon = llcrnrlon, llcrnrlat = llcrnrlat,
              urcrnrlat = urcrnrlat, urcrnrlon = urcrnrlon,
              area_thresh = 1000., resolution='i')
```

```python
##set resolution in km
x_resolution = 30
y_resolution = 30

##return a list of n size where n is the horizontal or vertical amount of grid spacings
##to get the desired resolution
##e.g., if you have a 100 km grid and you want your grid cells to be 10km each
##the length of xi and yi would be 10 or 100km / 10
xi,yi = get_grid(x_resolution,y_resolution,llcrnrlon,llcrnrlat,urcrnrlon,urcrnrlat)
mx,my = np.meshgrid(xi,yi)
mx,my = m(mx,my)

#this determines the distance decay rate of the density estimation function
dfactor = 100

##ckdtree procedure
zz = np.empty([len(yi),len(xi)])
zipped = zip(x, y)
kdtree = tree(zipped)
for xci in range(0, len(xi)):
    xc = xi[xci]
    for yci in range(0, len(yi)):
        yc = yi[yci]
        density = 0.
        retvalset = kdtree.query((xc,yc), k=5)
        for dist in retvalset[0]:
            density = density + math.exp(-dfactor * pow(dist, 2)) / 5
            zz[yci][xci] = min(density, 1.0) * 255

# create an axes on the right side of ax. The width of cax will be 5%
# of ax and the padding between cax and ax will be fixed at 0.05 inch.
zz_smooth = gf(zz,sigma=2,order=0)
zz_smooth[zz_smooth<20] = None
```

```python
num_contours = 10
plt.contourf(mx,my,zz_smooth,num_contours,cmap=plt.cm.get_cmap('Reds'),zorder=2)
plt.colorbar(shrink=.5)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawmapboundary(fill_color='gray')
m.fillcontinents(color='lightgray',lake_color='gray')
```

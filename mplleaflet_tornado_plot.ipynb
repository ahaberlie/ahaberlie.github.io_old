This script downloads tornado path shapefile information from SPCs dataset into your 
working directory and then unzips this information so it can be read by the program

If the shapefile is already in your working folder, the download will not occur

Although the SPC is trustworthy, this script uses an unzipping process so be
sure that the zip file you use is from a good source.

```python
import sys
import mplleaflet as mleaf
import matplotlib.pyplot as plt
import shapefile as sf
import pyproj
import numpy as np
import zipfile
import urllib
%matplotlib inline
```

You could modify the shapefile here but the code below it is based on the fields format presented at
http://www.spc.noaa.gov/wcm/SPC_severe_database_description.pdf

```python
try:
    #read in shapefile
    sfile = sf.Reader("tornado.shp")        
    print "Shapefile successfully read"
    sys.stdout.flush()
except:
    print "downloading tornado shapefile from SPC"
    sys.stdout.flush()
    urllib.urlretrieve ("http://www.spc.noaa.gov/gis/svrgis/zipped/tornado.zip", "tornado.zip")
    print "unzipping tornado shapefile"
    sys.stdout.flush()
    zf = zipfile.ZipFile('tornado.zip', 'r')
    zipfilelist = zf.namelist()
    
    for afile in zipfilelist:
        try:
            data = zf.read(afile)
        except KeyError:
            print 'ERROR: Did not find %s in zip file' % afile
            sys.stdout.flush()
        else:
            outfile = open(afile, 'wb')
            outfile.write(data)
            outfile.close()
    sfile = sf.Reader("tornado.shp")        
    print "Shapefile successfully read"
    sys.stdout.flush()
```

Next, we leverage pyshp (shapefile) to read in the geometry and record information

```python
##get all the shapes from your shapefile
shapes = np.array(sfile.shapes())

##get all the records for each shapefile
records = np.array(sfile.records())

##only those from 2013
shapes = shapes[records[:,1].astype(int) == 2013]
records = records[records[:,1].astype(int) == 2013]

print shapes[306].points
print records[306]
```

We then utilize pyproj to convert from "USA_Contiguous_Lambert_Conformal_Conic" to "WSG84"
mplleaflet uses lat/lon

```python
##the shapefiles are in USA_Contiguous_Lambert_Conformal_Conic
##find epsg code here: http://spatialreference.org/ref/epsg/wgs-84-utm-zone-16n/
p1 = pyproj.Proj(proj='lcc',lat_1=33,lat_2=45,lat_0=39,lon_0=-96,x_0=0,y_0=0,ellps='GRS80')

##want to convert points to WGS84 since this module takes lat/lon
##find epsg code here: http://spatialreference.org/ref/epsg/wgs-84/
p2 = pyproj.Proj(init='epsg:4326')
```

Next, we need to plot all of the paths in matplotlib.pyplot
mplleaflet basically exports your plot into a map, pretty much
bypassing basemap, for example.

```python
f, ax = plt.subplots()
##loop through all your shapes and records
for shape,record in zip(shapes,records):
    
    ##get a list of all the points for the current shape
    pts = np.array(shape.points[:])

    ##the 11th column is severity
    severity = record[10]
    
    ##transform from USACLCC to WGS84
    lon,lat = pyproj.transform(p1,p2,pts[:,0],pts[:,1])

    plt.hold(True)
    
    ##based on severity, assign different colors
    if severity=='0': 
       ax.plot(lon,lat,'b',label="EF0")
    elif severity=='1': 
       ax.plot(lon,lat,'g',label="EF1")
    elif severity=='2': 
       ax.plot(lon,lat,'c',label="EF2")
    elif severity=='3': 
       ax.plot(lon,lat,'y',label="EF3")
    elif severity=='4': 
       ax.plot(lon,lat,'m',label="EF4")
    elif severity=='5': 
       ax.plot(lon,lat,'r',label="EF5")
    else:
       ax.plot(lon,lat,'k')
       print "Unknown",severity
plt.title("Tornado path plots with no basemap, ready to export")
```

```python
##this writes out your html file and opens it automatically
##use mleaf.show() to output html file
mleaf.display(fig=ax.figure,tiles='mapbox bright')
```
